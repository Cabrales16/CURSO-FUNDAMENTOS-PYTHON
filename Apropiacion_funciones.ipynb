```python
total = 0

def alacancia(moneda):
  global total
  total = total + moneda
  return total

print(alacancia(10))
print(alacancia(20))
print(alacancia(5))
```

# Funciones recursivas

```python
def factorial_recursivo(n):
  if n == 1:
    return 1
  else:
    return n * factorial_recursivo(n-1)

num = int(input("ingrese le numero"))
print(f"el factorial de {num} es {factorial_recursivo(num)}")
```

```python
def atras(num):
  print(num)
  num -= 1
  if num > 0:
    atras(num)

atras(10)
```

1. Realiza una funci√≥n que tome una lista de n√∫meros enteros y devuelva dos listas ordenadas. La primera con los n√∫meros pares y la segunda con los n√∫meros impares.

```python
lista_num = [1,2,4,5,7,8,9,11,56,12,78,12,98,16,98,47,24,123,45,654,234,1]
pares = []
impares = []
def pares_impares():
  for i in lista_num:
    if i % 2 == 0:
      pares.append(i)
    else:
      impares.append(i)
  print(f'Los numeros pares son : {pares}')
  print(f'Los numeros impares son : {impares}')

pares_impares()
```

2. Realiza una funci√≥n que reciba una palabra y devuelva un diccionario donde las claves sean las diferentes letras que se encuentran en la palabra y los valores sean el n√∫mero de veces que se repite cada letra en la palabra

```python
palabra = input('Ingrese una palabra: ')
def contar_letras(palabra):
    diccionario = {}
    for letra in palabra:
        if letra in diccionario:
            diccionario[letra] += 1
        else:
            diccionario[letra] = 1
    return diccionario

resultado = contar_letras(palabra)
print(resultado)
```
