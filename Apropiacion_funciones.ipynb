```python
def cambio_valor():
  global x
  x = 10
x=8
cambio_valor()
print(x)
```

```python
total = 0

def alacancia(moneda):
  global total
  total = total + moneda
  return total

print(alacancia(10))
print(alacancia(20))
print(alacancia(5))
```

# Funciones recursivas

```python
def factorial_recursivo(n):
  if n == 1:
    return 1
  else:
    return n * factorial_recursivo(n-1)

num = int(input("ingrese le numero"))
print(f"el factorial de {num} es {factorial_recursivo(num)}")
```

```python
def atras(num):
  print(num)
  num -= 1
  if num > 0:
    atras(num)

atras(10)
```

1. Realiza una función que tome una lista de números enteros y devuelva dos listas ordenadas. La primera con los números pares y la segunda con los números impares.

```python
lista_num = [1,2,4,5,7,8,9,11,56,12,78,12,98,16,98,47,24,123,45,654,234,1]
pares = []
impares = []
def pares_impares():
  for i in lista_num:
    if i % 2 == 0:
      pares.append(i)
    else:
      impares.append(i)
  print(f'Los numeros pares son : {pares}')
  print(f'Los numeros impares son : {impares}')

pares_impares()
```

2. Realiza una función que reciba una palabra y devuelva un diccionario donde las claves sean las diferentes letras que se encuentran en la palabra y los valores sean el número de veces que se repite cada letra en la palabra

```python
palabra = input('Ingrese una palabra: ')
def contar_letras(palabra):
    diccionario = {}
    for letra in palabra:
        if letra in diccionario:
            diccionario[letra] += 1
        else:
            diccionario[letra] = 1
    return diccionario

resultado = contar_letras(palabra)
print(resultado)
```

3. Realiza una función que reciba un número y devuelva una cadena con los nombres de los números recibidos, separando cada número con un guión medio. Por ejemplo, si el número recibido es 134, la función devolverá la cadena "uno - tres - cuatro"

```python
numero = int(input("Ingrese un número: "))
lista_num = []
for i in str(numero):
  lista_num.append(int(i))
#print(lista_num)
dicc_numero = {1: 'uno', 2: 'dos', 3:'tres', 4: 'cuatro', 5:'cinco', 6:'seis', 7:'siete', 8: 'ocho', 9: 'nueve', 0: 'cero'}

def numero_letra ():
    final = " - ".join(dicc_numero[i] for i in lista_num) #Con el for toma cada elemento de la lista que este dentro del diccionario y el .join lo que hara es unirlos mediante " - "
    print(final)

numero_letra()
```
