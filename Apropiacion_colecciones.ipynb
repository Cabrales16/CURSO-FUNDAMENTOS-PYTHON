2. En una escuela se permite la inscripción de sus alumnos a las siguientes actividades deportivas: futbol (f), baloncesto (b). Realice un programa que permita ingresar el nombre de sus estudiantes e inscribir a cada uno en uno o los dos deportes.
La aplicación debe mostrar los siguientes listados:

*   Estudiantes inscritos en fulbol
*   Estudiantes inscritos en baloncesto
*   Estudiantes inscritos en ambos deportes
*   Todos los estudiantes inscritos en algún deporte
*   Estudiantes inscritos en solo un deporte.


Genere un menú con las opciones de la aplicación

```python
estudiantes = {"Douglas", "Natalia", "Cabrales", "Jhosep"}
futbol = []
baloncesto = []

print(f'Los estudiantes son: {estudiantes}')

print('Opciones:\n1)Agregar a futbol\n2)Agregar a baloncesto\n3)Agregar a ambos\n4)Finalizar')
while opcion > 0 and opcion < 5:
  opcion = int(input('Ingrese la opcion que desea realizar: '))
  if opcion != 4:
    estud = input('Seleccione el estudiante: ')

    if opcion == 1:
      print(f'El estudiante {estud} ha sido agregado a futbol')
      futbol.append(estud)

    elif opcion == 2:
      print(f'El estudiante {estud} ha sido agregado a baloncesto')
      baloncesto.append(estud)

    elif opcion == 3:
      print(f'El estudiante {estud} ha sido agregado a ambos')
      futbol.append(estud)
      baloncesto.append(estud)

    else:
      print('Estudiante no valido')

  else:
    print('El programa ha finzalido')
    break
colec_futbol = set(futbol)
colec_baloncesto = set(baloncesto)

print(f'Estudiantes inscritos en futbol: {colec_futbol}')
print(f'Estudiantes inscritos en baloncesto: {colec_baloncesto}')
print(f'Estudiantes inscritos en ambos: {colec_futbol.intersection(colec_baloncesto)}')
print(f'Estudiantes inscritos en un solo deporte: {colec_futbol.symmetric_difference(colec_baloncesto)}')
```

# Apropiación Diccionarios

1. Realice un programa traductor de frutas del español al inglés. Para ello, solicite el nombre de una fruta al usuario y muestre su traducción al inglés. Si la fruta no se encuentra en el diccionario, permítale al usuario agregarla con su respectiva traducción. El programa solo termina cuando el usuario ingresa una fruta que no está en el diccionario y no desea agregarla.

```python
frutas = {'manzana':'apple', 'naranja':'orange', 'platano':'banano', 'pera':'pear', 'coco': 'coconut', 'limon': 'lemon'}


while True:
  traduccionIngles = input("Ingrese el nombre de la fruta en español: ")
  if traduccionIngles in frutas:
    print(f"La traducción de {traduccionIngles} es: {frutas[traduccionIngles]}")
    break
  elif traduccionIngles not in frutas:
    print(f'La fruta {traduccionIngles} no se encuentra')
    añadir_res = input(f'Desea agregar la traducción de {traduccionIngles}?(si/no): ')
    if añadir_res == 'si':
      nueva_fruta = input(f'Cual es la traducción de {traduccionIngles}?: ')
      frutas[traduccionIngles] = nueva_fruta
    elif añadir_res == 'no':
        print('Volviendo al inicio...')
    else:
        print('Escoja una opción valida')
  else:
    print('Opcion no valida')
```
